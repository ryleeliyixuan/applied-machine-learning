## Load Data

```python
import matplotlib.pyplot as plt
import numpy as np

# Load the samples into an 150 × 4 (N x p) array
# where N is the number of samples and p is the number of attributes per sample
data = np.loadtxt("iris.data", delimiter=',', usecols=range(0,4));
  
print (data.shape);
# print (data[:5]);
```

```python
# create a N-dimensional vector containing each sample’s label (species)
vector = np.loadtxt("iris.data", delimiter=',', usecols=range(4,5), dtype=str);
print (vector.shape);
# print (vector[:5]);
```

## Visualization

```python
# map color from species
colorMap = {
    'Iris-setosa': "r",
    'Iris-versicolor': "g",
    'Iris-virginica': "b"
}
colors = list(map(lambda x: colorMap[x], vector));
# print (colors);
```

```python
# plot data points 
fig, axs = plt.subplots(4, 4, figsize=(12,12));

attributes = ["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"];

for i in range(len(attributes)):
    for j in range(len(attributes)):
        if i == j:
            axs[i, j].text(0, 0, attributes[i], horizontalalignment='left', verticalalignment='bottom');
        else:
            axs[j, i].scatter(data[:, i], data[:, j], c = colors, s = 7);

# save pics
plt.savefig ("hw0_iris_flower_plot.png");
```
